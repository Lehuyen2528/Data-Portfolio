```python
# what companies do developer get the paid the most?
# Grouping the data frame by "OrgSize" and calculating the average "ConvertedCompYearly"
avg_comp_by_org_size = data.groupby("OrgSize")["ConvertedCompYearly"].mean()

# Finding the organization size with the highest average compensation
highest_paid_org_size = avg_comp_by_org_size.idxmax()
print("The organization size  paid the most is:", highest_paid_org_size)
```

```python
# How much does remote working matter to employees?
data.RemoteWork.value_counts().plot(kind = 'bar')
```

```python
# How does coding experience affect the level of pay?
data.groupby('WorkExp')['ConvertedCompYearly'].mean().plot(kind = 'bar')
```

```python
# What’s the most popular method of learning to code?
data.LearnCode.value_counts()[:10]
```

```python
# Are you more likely to get a job as a developer if you have a master’s degree?
employment_and_Edlevel = data.groupby(['EdLevel','Employment']).size().unstack()
sns.heatmap(employment_and_Edlevel, cmap= sns.cubehelix_palette(as_cmap= True))
```
