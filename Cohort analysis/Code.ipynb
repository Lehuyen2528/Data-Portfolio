### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

### Load dataset

```python
data = pd.read_excel('Online Retail.xlsx')
data.info()
```

```python
#take a look at the head 
data.head()
```

```python
# Drop row with no customer ID

data = data.dropna(subset = ['CustomerID'])
data.info()
```

```python
# Create an invoice month
def get_month(x):
    return dt.datetime(x.year,x.month,1)

data['InvoiceMonth'] = data['InvoiceDate'].apply(get_month)
data.tail()
```

```python
# Create a column index with the minimum invoice date first time customer was acuired
data['Cohort Month'] = data.groupby('CustomerID')['InvoiceMonth'].transform('min')
data.head(10)
```
